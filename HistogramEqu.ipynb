```python
import cv2
import matplotlib.pyplot as plt
from google.colab import files
import numpy as np
import io
```

```python
uploaded = files.upload()
```

```python
file_name_1 = 'soru1-chest.tif'
```

```python
image = cv2.imdecode(np.frombuffer(uploaded[file_name_1], np.uint8), cv2.IMREAD_GRAYSCALE)
```

```python
histogram_1 = cv2.calcHist([image], [0], None, [256], [0, 256])
```

```python
plt.figure()
plt.title('Chest Histogram')
plt.xlabel('Pixel Değeri')
plt.ylabel('Pixel Sayısı')
plt.plot(histogram_1)
plt.xlim([0, 256])
plt.show()
```
